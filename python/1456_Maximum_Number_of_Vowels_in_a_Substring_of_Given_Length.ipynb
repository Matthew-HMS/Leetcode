```python
class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        vowels = {'a', 'e', 'i', 'o', 'u'}
        counter = 0
        for i in range(k):
            if s[i] in vowels:
                counter += 1

        max_counter = counter
        for i in range(k, len(s)):
            if s[i] in vowels:
                counter += 1
            if s[i-k] in vowels:
                counter -= 1

            max_counter = max(counter, max_counter)

        return max_counter
```

```python
class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        vowels = {'a', 'e', 'i', 'o', 'u'}
        ans = cnt = 0
        for i, c in enumerate(s):
            if c in vowels:
                cnt += 1
            if i >= k and s[i - k] in vowels:
                cnt -= 1
            ans  = max(cnt, ans)
        return ans    
```
