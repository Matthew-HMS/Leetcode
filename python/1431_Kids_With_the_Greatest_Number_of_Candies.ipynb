```python
from typing import List

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_num = -1
        for i in candies:
            if i > max_num:
                max_num = i

        base = max_num - extraCandies

        output = []
        for i in candies:
            if i >= base:
                output.append(True)
            else:
                output.append(False)

        return output
    
s = Solution()
ans = s.kidsWithCandies([2,3,5,1,3], 3)
print(ans)
```

```python
from typing import List

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_num = max(candies)

        base = max_num - extraCandies
        output = [i >= base for i in candies]
    
        return output
    
s = Solution()
ans = s.kidsWithCandies([2,3,5,1,3], 3)
print(ans)
```
