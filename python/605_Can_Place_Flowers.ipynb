```python
class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        counter = 0
        i = 0
        if len(flowerbed) == 1:
            if flowerbed[0] == 0:
                counter += 1
            return counter >= n
            
        while i < len(flowerbed) - 1:
            if flowerbed[i] == 0 and flowerbed[i+1] == 0:
                flowerbed[i] = 1
                counter += 1
                i += 2

            elif flowerbed[i] == 1:
                i += 2

            else:
                i += 1

        if len(flowerbed) >= 3:
            if flowerbed[-3] == 1 and flowerbed[-2] == 0 and flowerbed[-1] == 0:
                counter += 1

        return counter >= n
```
