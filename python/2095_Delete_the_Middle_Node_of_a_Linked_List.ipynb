```python
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        middle = head
        temp = head
        counter = 0
        while temp:
            temp = temp.next
            counter += 1
            if counter % 2 == 0:
                prev = middle
                middle = middle.next

        if counter == 1:
            return None
        prev.next = prev.next.next
        middle.next = None

        return head
```

```python
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return None
        prev, middle, temp = None, head, head

        while temp and temp.next:
            prev = middle
            middle = middle.next
            temp = temp.next.next

        prev.next = prev.next.next
        middle.next = None

        return head
```
