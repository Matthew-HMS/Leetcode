Time: O(N), Space: O(1)

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        output = [1]
        prefix = 1
        suffix = 1
        for i in range(1, len(nums)):
            prefix = prefix * nums[i-1]
            output.append(prefix)

        for i in range(len(nums) - 2, -1, -1):
            suffix = suffix * nums[i+1]
            output[i] *= suffix
        
        return output
```
