```python
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        i = 0
        write = 0
        while i < len(nums):
            if nums[i] != 0:
                nums[write] = nums[i]
                write += 1
            i += 1

        while write < len(nums):
            nums[write] = 0
            write += 1
```
