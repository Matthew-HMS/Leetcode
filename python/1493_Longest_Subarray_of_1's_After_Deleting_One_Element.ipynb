```python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        max_len = 0
        i, j, zero = 0, 0, 0
        counter = 0
        while j < len(nums):
            if nums[j] == 0:
                if counter == 0:
                    counter += 1
                else:
                    max_len = max(max_len, j-i)
                    i = zero + 1
                zero = j
            
            j += 1

        max_len = max(max_len, j-i)
        return max_len - 1
```

```python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        left = 0
        zeros = 0
        max_len = 0
        
        for i in range(len(nums)):
            if nums[i] == 0:
                zeros += 1
            
            while zeros > 1:
                if nums[left] == 0:
                    zeros -= 1
                left += 1

            max_len = max(max_len, i-left)
        
        return max_len
```
