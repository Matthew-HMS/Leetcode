```python
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        left, right = 0, 0
        counter = 0
        max_window = 0
        while right < len(nums):
            if nums[right] == 0:
                if counter < k:
                    counter += 1
                else:
                    window = right - left
                    max_window = max(window, max_window)
                    while nums[left] == 1:
                        left += 1
                    else:
                        left += 1

            right += 1

        window = right - left
        max_window = max(window, max_window)
        
        return max_window
```

```python
class Solution:
    def longestOnes(self, nums: List[int], k: int) -> int:
        left, right = 0, 0
        counter = 0
        max_window = 0
        while right < len(nums):
            if nums[right] == 0:
                counter += 1
            
            while counter > k:
                if nums[left] == 0:
                    counter -= 1
                left += 1

            window = right - left + 1
            max_window = max(window, max_window)

            right += 1

        return max_window
```
