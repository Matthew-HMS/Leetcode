Time: O(n), Space: O(n)

```python
class Solution:
    def pairSum(self, head: Optional[ListNode]) -> int:
        val = []
        while head:
            val.append(head.val)
            head = head.next
        l, r = 0, len(val)-1
        max_sum = -1
        while l < r:
            max_sum = max(max_sum, val[l] + val[r])
            l += 1
            r -= 1
        return max_sum
```
