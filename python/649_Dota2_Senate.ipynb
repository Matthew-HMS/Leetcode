```python
from collections import deque
class Solution:
    def predictPartyVictory(self, senate: str) -> str:
        r_q = deque()
        d_q = deque()

        for i in range(len(senate)):
            if senate[i] == "R":
                r_q.append(i)
            else:
                d_q.append(i)

        while r_q and d_q:
            if r_q[0] < d_q[0]:
                d_q.popleft()
                r_q.append(r_q.popleft() + len(senate))
            else:
                r_q.popleft()
                d_q.append(d_q.popleft() + len(senate))

        return "Radiant" if r_q else "Dire"
```
