```python
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        prev = None
        curr = None
        nxt = head

        while nxt != None:
            prev = curr
            curr = nxt
            nxt = curr.next
            curr.next = prev

        head = curr
        return head
```
