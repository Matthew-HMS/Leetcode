Space: O(n)

```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        if len(cost) == 2:
            return min(cost[0], cost[1])
        steps = [cost[0], cost[1]]
        for i in range(2, len(cost)):
            step_cost = min(steps[i-1], steps[i-2]) + cost[i]
            steps.append(step_cost)

        return min(steps[-1], steps[-2])
```

Space: O(1)

```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        a, b = cost[0], cost[1]
        if len(cost) == 2:
            return min(a, b)
        for i in range(2, len(cost)):
            step_cost = min(a, b) + cost[i]
            a, b = b, step_cost

        return min(a, b)
```

Do it from back

```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        cost.append(0) # top [10,15,20,0]
        
        for i in range(len(cost) - 4, -1, -1):
            cost[i] += min(cost[i+1], cost[i+2])
            
        return min(cost[0], cost[1])
```
