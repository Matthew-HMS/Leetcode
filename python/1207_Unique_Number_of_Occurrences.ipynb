```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        map = {}
        for i in range(len(arr)):
            map[arr[i]] = map.get(arr[i], 0) + 1

        val = {}
        for value in map.values():
            val[value] = val.get(value, 0) + 1
        
        for value in val.values():
            if value != 1:
                return False
        
        return True
```

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        map = {}
        for i in range(len(arr)):
            map[arr[i]] = map.get(arr[i], 0) + 1

        occur = []
        for value in map.values():
            occur.append(value)

        if len(occur) == len(set(occur)):
            return True
        else:
            return False
```

```python
class Solution:
    def uniqueOccurrences(self, arr: List[int]) -> bool:
        freq = {}
        for x in arr:
            freq[x] = freq.get(x, 0) + 1

        return len(freq) == len(set(freq.values()))
```
