```python
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        ans = nums[0]
        for i in range(1, len(nums)):
            ans = ans ^ nums[i]

        return ans
```

Follow up: what if not two, but each number appear three times? Find the one that appears only once.

```python
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        ans = 0
        for i in range(32):
            count = 0
            for x in nums:
                if (x >> i) & 1:
                    count += 1
            
            # If count is not a multiple of 3, the unique number has this bit
            if count % 3 != 0:
                # Handle Python's infinite precision integers for negative numbers
                if i == 31: # The sign bit
                    ans -= (1 << i)
                else:
                    ans |= (1 << i)
        return ans
```
