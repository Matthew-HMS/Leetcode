```python
class Solution:
    def maxArea(self, heights: List[int]) -> int:
        max_v = 0
        left, right = 0, len(heights) - 1
        while left < right:
            v = min(heights[left], heights[right]) * (right - left)
            max_v = max(max_v, v)
            if heights[left] <= heights[right]:
                left += 1
            else:
                right -= 1

        return max_v
```
