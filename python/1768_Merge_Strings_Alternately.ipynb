```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """
        result = []
        i = 0
        while i < len(word1) or i < len(word2):
            if i < len(word1):
                result.append(word1[i])
            if i < len(word2):
                result.append(word2[i])
            i += 1
        return ''.join(result)
    
s = Solution()
ans = s.mergeAlternately("abc", "pqrst")
print(ans)
```

if merge K strings, instead of 2

```python
import itertools

class Solution:
    def mergeKStrings(self, words: list[str]) -> str:
        output = []
        for triplet in itertools.zip_longest(*words):
            # triplet looks like ('a', 'p', None)
            # add only the valid characters
            for ch in triplet:
                if ch is not None:
                    output.append(ch)
            
        return "".join(output)
    
s = Solution()
ans = s.mergeKStrings(["abc", "pqrst", "uvwiol"])
print(ans)
```

```python
from itertools import zip_longest

class Solution:
    def mergeAlternately(self, word1: str, word2: str) -> str:
        # zip_longest pairs characters and uses an empty string as a fillvalue
        # for whichever string is shorter.
        zipped = zip_longest(word1, word2, fillvalue='')
        
        # We use a generator expression inside join to keep memory usage low
        return "".join(char for pair in zipped for char in pair)
```
