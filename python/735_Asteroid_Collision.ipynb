```python
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        stack = []
        for n in asteroids:
            if n > 0:
                stack.append(n)
            else:
                while len(stack) > 0 and stack[-1] < abs(n) and stack[-1] > 0:
                    stack.pop()
                if len(stack) > 0 and stack[-1] == abs(n) and stack[-1] > 0:
                    stack.pop()
                elif len(stack) == 0 or stack[-1] < 0:
                    stack.append(n)
                    
        return stack
```

```python
class Solution:
    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        
        res = []

        for a in asteroids:

            while res and a < 0 < res[-1]:
                if -a > res[-1]:
                    res.pop()
                    continue
                elif -a == res[-1]:
                    res.pop()
                break
            else:
                res.append(a)

        return res
```
