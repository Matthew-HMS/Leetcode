Time: O(nlogn)

```python
class Solution:
    def countBits(self, n: int) -> List[int]:
        ans = []
        for i in range(n+1):
            counter = 0
            while i:
                i &= (i-1)
                counter += 1

            ans.append(counter)

        return ans
```

DP way, Time: O(n)

```python
class Solution:
    def countBits(self, n: int) -> List[int]:
        ans = [0] * (n+1)
        for i in range(n+1):
            ans[i] = ans[i >> 1] + (i & 1)
        
        return ans
```
