Time: O(nlogn), Space: O(1), due to sorting method

```python
class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        nums = sorted(nums)
        left, right = 0,  len(nums)-1
        counter = 0
        while left < right:
            s = nums[left] + nums[right]
            if s < k:
                left += 1
            elif s > k:
                right -= 1
            elif s == k:
                counter += 1
                left += 1
                right -= 1

        return counter
```

Time: O(n), Space: O(n)

```python
from collections import Counter
class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        counter = 0
        m = Counter(nums)
        for key in m.keys():
            if m[k-key]:
                if (k-key) != key:
                    n = min(m[key], m[k-key])
                    counter += n
                    m[key] -= n
                    m[k-key] -= n
                elif (k-key) == key and m[key] >= 2:
                    counter += (m[key] // 2)

        return counter
```
