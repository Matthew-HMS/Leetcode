```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        prefix = 0
        suffix = sum(nums) - nums[0]
        if prefix == suffix:
            return 0
        for i in range(1, len(nums)):
            prefix += nums[i-1]
            suffix -= nums[i]
            
            if prefix == suffix:
                return i
            
        
        return -1
```

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        left_sum = 0
        total_sum = sum(nums)
        
        for i, x in enumerate(nums):
            # The right sum is: total_sum - left_sum - current_element
            if left_sum == (total_sum - left_sum - x):
                return i
            left_sum += x
            
        return -1
```
