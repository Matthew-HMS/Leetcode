```python
class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        t_p = 0
        for i in range(len(s)):
            while t_p < len(t) and t[t_p] != s[i]:
                t_p += 1
            if t_p >= len(t):
                return False

            t_p += 1

        return True
```
